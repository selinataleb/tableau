```python
#Import Dependencies

import pandas as pd
import numpy as np
```

```python
july1_df = pd.read_csv("../Ressources /2018-citibike-tripdata/Summer/7_July/201807-citibike-tripdata_1.csv")
july2_df = pd.read_csv("../Ressources /2018-citibike-tripdata/Summer/7_July/201807-citibike-tripdata_2.csv")
```

```python
# Concatenate the DataFrames
july_df = pd.concat([july1_df, july2_df], ignore_index=True)
july_df.info()
```

```python
#checking for null values 
july_df.isnull().values.any()
```

```python
#checking for january 

january_df = pd.read_csv("../Ressources /2018-citibike-tripdata/Winter/1_January/201801-citibike-tripdata.csv")
january_df.head()
```

```python
#checking for null values 
january_df.isnull().values.any()
```

```python
#merging both july and january into one fle 

juljan_df = pd.concat([july_df, january_df], ignore_index=True)
juljan_df.info()
```

```python
#changing birth year to age 

current_year = 2018 
juljan_df['age'] = current_year - juljan_df['birth year']
juljan_df['age']

```

```python
#dropping birth year column and it is being replaced by age 
juljan_df = juljan_df.drop(columns = ["birth year"])
juljan_df.head()
```

```python
#dropping the unknown type of gender to remove uncertainty in our analysis
juljan_df = juljan_df[juljan_df['gender'] != 0]
```

```python
#replacing numerical values in the gender column by their assigned roles (1 = Male, 2= Female)

juljan_df['gender']= juljan_df['gender'].astype(int)

juljan_df['gender'] = juljan_df['gender'].replace({1: 'Male', 2:'Female'})
```

```python
juljan_df.head()
```

```python
#checking for age column data type 

juljan_df['age'] = juljan_df['age'].astype(int)  
```

```python
#drop duplicates 

juljan_df = juljan_df.drop_duplicates(keep = 'first')
```

```python
juljan_df.info()
```

```python
juljan_df.reset_index(drop=True, inplace=True)
```

```python
juljan_df.info()
```

```python
# Convert 'tripduration' to numeric with coercion for non-numeric values (I used chatgpt for this one because I was getting errors)
juljan_df['tripduration'] = pd.to_numeric(juljan_df['tripduration'], errors='coerce')

# Drop rows with NaN in 'tripduration' after conversion
juljan_df.dropna(subset=['tripduration'], inplace=True)

# Convert 'tripduration' to integer 
juljan_df['tripduration'] = juljan_df['tripduration'].astype(int)

# Filter for values less than or equal to 57600
juljan_df = juljan_df[juljan_df['tripduration'] <= 57600].reset_index(drop=True)


juljan_df.info()
```

```python
juljan_df.min()
```

```python
juljan_df = juljan_df[juljan_df['age'] <= 55].reset_index(drop=True)
juljan_df.info()
```

```python
#exporting the file 

juljan_df.to_csv('juljanbikee.csv', index=False)
```

# Brief code explanation 

For any questions we want to tackle regarding gender, we dropped the Unknown values that were equal 0 and it dropped by 300k thus anything related to gender is minimized from uncertainty. Our gender column is reduced to 2 variables: Male and Female. 

For any questions related to age, we changed the birthyear column to age by calculating the difference of the current year (which as per the dataset is 2018) and the birthyear. We have capped our age dataset to 55 years old because after having done research I found that the majority of riders fall between the ages of 25-55. 
Source Used: https://nycdatascience.com/blog/student-works/data-study-on-citi-bike-riders-in-different-ages/

Our target for the purpose of this project is the day pass users so we will be limited our trip durations to 16 hours a day or in seconds as the dataset presents it in seconds would be 57600 seconds.
